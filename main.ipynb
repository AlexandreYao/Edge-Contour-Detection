```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Open the image
image_name = r"image.jpg"
img = cv2.imread(image_name)
edges = cv2.Canny(img, 100, 200, 3, L2gradient=True)
# Create a subplot with two columns and one row
plt.figure(figsize=(10, 5))  # Adjust the figure size as needed
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis("off")  # Hide axis
plt.title(f"{image_name}") #
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")
plt.axis("off")  # Hide axis
plt.title(f"{image_name} canny")
plt.show()
```

```python
# Load the original image
image_name = r"phones.jpg"
image = cv2.imread(image_name)

# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find Canny edges
edged = cv2.Canny(gray, 30, 200)

# Finding Contours
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Draw all contours on a copy of the original image
image_with_contours = image.copy()
cv2.drawContours(image_with_contours, contours, -1, (0, 255, 0), 3)

# Plotting with matplotlib
plt.figure(figsize=(20, 30))

# Plot the original image in the first column
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Plot the image with contours in the second column
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_with_contours, cv2.COLOR_BGR2RGB))
plt.title('Image with Contours')
plt.axis('off')

plt.show()
```
